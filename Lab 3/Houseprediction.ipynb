# Lab 3: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
import pandas as pd
```

```python
df = pd.read_csv("House Price Prediction Dataset.csv")
```

```python
print("Dataset Shape (rows, columns):")
print(df.shape)
```

```python
print("\nFirst 5 Rows:")
print(df.head())
```

```python
print("\nColumn Names:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Price'], bins=50)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

The House Prices show a right-skewed distribution, indicating that most houses fall within a lower price range with a few  high-value outliers.

```python
df['Price'].describe()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include=['number']).corr(), 
            annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```

Area and number of bedrooms show how strong positive correlation with house price, suggesting they are important predictive features.

```python
plt.scatter(df['Area'], df['Price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

There is a clear positive relationship between area and price.

```python
sns.boxplot(x=df['Price'])
plt.title("Price Boxplot")
plt.show()
```
